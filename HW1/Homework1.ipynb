1. Substitution Cipher:

A substitution cipher is a type of encryption where each letter in the plaintext is replaced with another letter or symbol.

Differences from Shift Cipher:

While the shift cipher is a specific type of substitution cipher (Caesar cipher), a general substitution cipher can replace each letter with any other, providing more complexity.

Creation with Keyword or Permutation:

You can create a substitution cipher by creating a random permutation of the alphabet or using a keyword to generate a key for the substitution. The keyword method involves aligning the keyword with the alphabet and then using the resulting permutation.

Advantages and Disadvantages:

- Advantages: Provides a higher level of security compared to a shift cipher. The use of keywords makes it adaptable and allows for a larger key space.
- Disadvantages: Vulnerable to frequency analysis, especially if the language structure is retained. The key size can be large, making key distribution a challenge.

Breaking with Frequency Analysis:

Frequency analysis involves analyzing the frequency of letters in the ciphertext. Letters that occur most frequently are likely to represent common letters in the language, helping to deduce the substitutions.

Historical Example:

The Caesar cipher, a type of substitution cipher, was used by Julius Caesar to communicate securely with his generals.

2. Affine Cipher:

The affine cipher combines elements of both the shift and substitution ciphers by using modular arithmetic to encrypt and decrypt messages.

Encryption and Decryption:

The affine cipher encrypts a letter using the mathematical formula $E(x)=(ax+b)$ mod m, where a and b are key values, and m is the size of the alphabet.

Advantages and Vulnerabilities:

- Advantages: Introduces complexity compared to basic shift ciphers. The use of modular arithmetic enhances security.

- Vulnerabilities: Vulnerable to known plaintext attacks, especially if the attacker knows two corresponding plaintext and ciphertext pairs.

Breaking with Known Plaintext Attacks:

If an attacker has access to both the plaintext and corresponding ciphertext, they can deduce the key values using mathematical analysis.

Practical Use:

The affine cipher can be used in situations where a balance between complexity and ease of implementation is required.

3. Hill Cipher:

The Hill cipher is a polygraphic substitution cipher that uses linear algebra concepts, involving matrix multiplication, to encrypt and decrypt messages.

Encryption and Decryption:

The Hill cipher encrypts blocks of letters by multiplying them with a key matrix. Decryption involves the multiplication of the ciphertext with the modular inverse of the key matrix.

Advantages and Limitations:

- Advantages: Provides a higher level of security due to its matrix-based approach. Key distribution is easier compared to other classical ciphers.
- Limitations: Vulnerable to attacks when the key matrix is not chosen properly. Larger key sizes may pose implementation challenges.

Matrix Operations:

The Hill cipher uses matrix multiplication and modular arithmetic, making it resistant to basic frequency analysis.

Real-World Application:

The Hill cipher has been applied in secure communication systems where the use of matrix operations is feasible.

4. Permutation Cipher:

The permutation cipher, also known as a transposition cipher, rearranges the order of the letters in the plaintext to create the ciphertext.

Encryption and Decryption:

The permutation cipher encrypts by rearranging the order of the letters according to a specific key. Decryption involves reversing this process.

Strengths and Weaknesses:

- Strengths: Provides security by disrupting the linear structure of the plaintext. Different key permutations yield different ciphertexts for the same plaintext.
- Weaknesses: Vulnerable to brute force attacks, especially if the key length is short. The security heavily relies on the key permutation.

Breaking with Brute Force or Frequency Analysis:

Brute force involves trying all possible permutations until the correct one is found. Frequency analysis is less effective but may reveal patterns in specific cases.

Historical Example:

The Spartan military used a form of transposition cipher known as the Scytale during ancient times.

Write a Python function substitution_cipher_encrypt(plaintext, key) that takes a plaintext string and a substitution key as input and returns the corresponding ciphertext using the substitution cipher.

```python
import math
import numpy as np
from collections import Counter
# Substitution Cipher
def substitution_cipher_encrypt(plaintext, key):
    ciphertext = ''.join(key.get(char, char) for char in plaintext)
    return ciphertext

Plaintext= "HELLO"
Key= {'H': 'X', 'E': 'Y', 'L': 'Z', 'O': 'A'}
#Ciphertext: "XYZZA"

print(substitution_cipher_encrypt(Plaintext, Key))
```

Exercise:
Write a Python function affine_cipher_encrypt(plaintext, a, b) that takes a plaintext string and two integer values a and b as input and returns the corresponding ciphertext using the affine cipher.

```python
# Affine Cipher
def affine_cipher_encrypt(plaintext, a, b):
    ciphertext = ''.join(chr((a * (ord(char) - ord('A')) + b) % 26 + ord('A')) if char.isalpha() else char for char in plaintext.upper())
    return ciphertext

plaintext= "HELLO"
a= 5
b= 8

print(affine_cipher_encrypt(plaintext, a, b))

#Ciphertext: "RCLLA"
```

3. Hill Cipher
Exercise:
Write a Python function hill_cipher_encrypt(plaintext, key) that takes a plaintext string and a 2x2 matrix key as input and returns the corresponding ciphertext using the Hill cipher.

```python
# Hill Cipher
def hill_cipher_encrypt(plaintext, key):
    n = len(key)
    plaintext = [ord(char) - ord('A') for char in plaintext.upper() if char.isalpha()]
    ciphertext = ''
    for i in range(0, len(plaintext), n):
        block = plaintext[i:i+n]
        result = [(sum(row[i] * block[j] for i, j in enumerate(range(len(block)))) % 26) + ord('A') for row in key]
        ciphertext += ''.join(chr(char) for char in result)
    return ciphertext

Plaintext= "HELLOWORLD"
Key= [[2, 3], [1, 4]]
#Ciphertext: "AXDDQYBEFX"

print(hill_cipher_encrypt(Plaintext, Key))
```

4. Permutation Cipher

Write a Python function permutation_cipher_encrypt(plaintext, key) that takes a plaintext string and a permutation key as input and returns the corresponding ciphertext using the permutation cipher.

```python
# Permutation Cipher
def permutation_cipher_encrypt(plaintext, key):
    ciphertext = ''.join(plaintext[i - 1] for i in key)
    return ciphertext

Plaintext= "HELLO"
Key= [3, 1, 4, 2, 5]
#Ciphertext: "LHLEO"

print(permutation_cipher_encrypt(Plaintext, Key))
```

Exercise 2
Consider a cryptosystem in which the sets representing the laintext, ciphertext and keys are: $P=a,b,c, K = K_1,K_2,K_3$ and $C=1,2,3,4$ Suppose the encryption matrix is as follows:

| | a| b | c|
| ----- | --- | ----- |----|
|K1|	1	|2|	3|
|K2|	2	|3|	4|
|K3|	3|	4|	1|

Given that keys are chosen equiprobably, and the plaintext probability distribution is $Pr[a] = \frac{1}{2}, Pr[b] =\frac{1}{3}, Pr[c]=\frac{1}{6}$, compute $H(P), H(C), H(K), H(C|K)$ and $H(P|C)$

```python
# Function to calculate entropy
def entropy(probabilities):
    return -sum(p * math.log2(p) for p in probabilities if p > 0)

# Given probabilities for plaintext (P), ciphertext (C), and keys (K)
prob_P = [1/2, 1/3, 1/6]
prob_C = [1/4, 1/4, 1/4, 1/4]
prob_K = [1/3, 1/3, 1/3]

# Compute entropies
H_P = entropy(prob_P)
H_C = entropy(prob_C)
H_K = entropy(prob_K)

# Joint entropy of C and K
H_CK = H_C + H_K

# Conditional entropy of C given K
H_C_given_K = H_CK - H_K

# Conditional entropy of P given C
H_P_given_C = H_P - H_C

# Print the results
print(f"H(P) = {H_P:.4f}")
print(f"H(C) = {H_C:.4f}")
print(f"H(K) = {H_K:.4f}")
print(f"H(C|K) = {H_C_given_K:.4f}")
print(f"H(P|C) = {H_P_given_C:.4f}")
```
